## 1. **Emotion-based Voice Modulation**

Implementation Steps:
Based on the emotion, adjust parameters such as pitch, speed, and tone.
For instance:

Happy: Higher pitch, faster speed.

Sad: Lower pitch, slower speed.

Angry: Faster speech, higher pitch.

```python
from TTS.api import TTS

# Initialize TTS synthesizer
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC")  # Remove use_cuda


# Function to adjust speech based on emotion
def adjust_speech_based_on_emotion(text, emotion):
    # Set default values for pitch and speed
    if emotion == "happy":
        pitch = 1.2  # Increase pitch for happy emotion
        speed = 1.3  # Faster speech for happiness
    elif emotion == "sad":
        pitch = 0.7  # Lower pitch for sadness
        speed = 0.7  # Slower speech for sadness
    elif emotion == "angry":
        pitch = 2.5  # Slightly higher pitch for anger
        speed = 1.2  # Faster speech for anger
    else:
        pitch = 1.0  # Neutral pitch
        speed = 1.0  # Normal speed

    # Generate audio with adjustments
    audio_path = f"{emotion}_output.wav"
    tts.tts_to_file(
        text=text,
        file_path=audio_path,
        speaker_idx=None,  # You can specify a speaker if the model supports it
        speed=speed,       # Adjust speed
        emotion=None       # If the model supports emotions, set it here
    )
    print(f"Audio with {emotion} emotion saved to {audio_path}")

# Example usage
text = "Hello, how are you doing today?, is everything going fine"
emotion = "angry"  # You can change to 'sad', 'angry', or 'neutral'
adjust_speech_based_on_emotion(text, emotion)
```

```python
import os
import time
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
from playsound import playsound

# Initialize components
recognizer = sr.Recognizer()
translator = Translator()

def recognize_speech_from_mic():
    """Captures speech from the microphone and converts it to text."""
    with sr.Microphone() as source:
        print("Listening...")
        try:
            audio = recognizer.listen(source, timeout=5)  # Adjust timeout as needed
            print("Processing...")
            text = recognizer.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            return "Sorry, I could not understand the audio."
        except sr.RequestError as e:
            return f"Error with the recognition service: {e}"
        except Exception as e:
            return f"Error: {e}"

def translate_text(input_text, target_language):
    """Translates text to the target language."""
    translated = translator.translate(input_text, dest=target_language)
    return translated.text

def text_to_speech(text, lang, output_file):
    """Converts text to speech and saves it to an audio file."""
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)

def live_voice_to_voice_translation(source_language, target_language):
    """Live voice-to-voice translation system."""
    try:
        while True:
            # Recognize speech
            input_text = recognize_speech_from_mic()
            if not input_text:
                print("No input detected. Retrying...")
                continue

            print(f"Recognized Speech: {input_text}")

            # Translate text
            print(f"Translating from {source_language} to {target_language}...")
            translated_text = translate_text(input_text, target_language)
            print(f"Translated Text: {translated_text}")

            # Generate speech
            print("Generating speech...")
            output_audio = f"output_{target_language}.mp3"
            text_to_speech(translated_text, target_language, output_audio)

            # Play the audio
            print("Playing audio...")
            playsound(output_audio)

            # Clean up
            os.remove(output_audio)

            # Optional: Break loop (press Ctrl+C or condition)
    except KeyboardInterrupt:
        print("Translation stopped.")
    except Exception as e:
        print(f"Error: {e}")

# Example usage
if __name__ == "__main__":
    source_lang = "en"  # English (recognition)
    target_lang = "es"  # Spanish (translation and synthesis)
    live_voice_to_voice_translation(source_lang, target_lang)
```

```python
import os
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display
import speech_recognition as sr

# Initialize Translator and Recognizer
translator = Translator()
recognizer = sr.Recognizer()

def transcribe_audio(file_path):
    """Transcribes audio file into text."""
    with sr.AudioFile(file_path) as source:
        audio_data = recognizer.record(source)
        return recognizer.recognize_google(audio_data)

def translate_text(input_text, target_language):
    """Translates text to the specified target language."""
    translated = translator.translate(input_text, dest=target_language)
    return translated.text

def text_to_speech(text, lang, output_file):
    """Converts translated text to speech and saves it as an audio file."""
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)

def process_translation(input_audio_path, target_language):
    """Handles the workflow for live voice-to-voice translation."""
    if not os.path.exists(input_audio_path):
        print(f"Error: File '{input_audio_path}' not found.")
        return

    # Step 1: Transcribe Audio
    print("Step 1: Transcribing audio...")
    try:
        original_text = transcribe_audio(input_audio_path)
        print(f"Original Text: {original_text}")
    except Exception as e:
        print(f"Error during transcription: {e}")
        return

    # Step 2: Translate Text
    print(f"\nStep 2: Translating text to {target_language}...")
    try:
        translated_text = translate_text(original_text, target_language)
        print(f"Translated Text: {translated_text}")
    except Exception as e:
        print(f"Error during translation: {e}")
        return

    # Step 3: Text-to-Speech Conversion
    print("\nStep 3: Generating audio from translated text...")
    try:
        output_audio = "translated_audio.mp3"
        text_to_speech(translated_text, target_language, output_audio)
        print(f"Audio file '{output_audio}' generated successfully.")
    except Exception as e:
        print(f"Error during text-to-speech conversion: {e}")
        return

    # Step 4: Play the Audio
    print("\nStep 4: Playing translated audio...")
    display(Audio(output_audio, autoplay=True))

# Example usage
# Provide the path to your .wav file and target language code (e.g., 'es' for Spanish)
input_audio_path = input("Enter the full path to your .wav audio file: ").strip()
target_lang = input("Enter the target language code (e.g., 'es' for Spanish): ").strip()
process_translation(input_audio_path, target_lang)
```

```python
import os
import io
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display
import speech_recognition as sr

# Initialize Translator and Recognizer
translator = Translator()
recognizer = sr.Recognizer()

def get_audio_file():
    """Prompts the user to specify the path to their audio file."""
    print("Please specify the full path to your .wav audio file (e.g., /path/to/audio.wav):")
    file_path = input("Audio file path: ").strip()
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    return file_path

def transcribe_audio(file_path):
    """Transcribes audio file into text."""
    with sr.AudioFile(file_path) as source:
        print("Processing audio file for transcription...")
        audio_data = recognizer.record(source)
        return recognizer.recognize_google(audio_data)

def translate_text(input_text, target_language):
    """Translates text to the specified target language."""
    print(f"Translating text to language: {target_language}")
    translated = translator.translate(input_text, dest=target_language)
    return translated.text

def text_to_speech(text, lang, output_file):
    """Converts translated text to speech and saves it as an audio file."""
    print("Generating speech from the translated text...")
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)
    print(f"Audio saved as: {output_file}")

def process_translation(target_language):
    """Handles the workflow for live voice-to-voice translation."""
    # Step 1: Get the audio file path
    try:
        input_file = get_audio_file()
    except FileNotFoundError as e:
        print(e)
        return

    # Step 2: Transcribe Audio
    print("\nStep 2: Transcribing audio...")
    try:
        original_text = transcribe_audio(input_file)
        print(f"Original Text: {original_text}")
    except Exception as e:
        print(f"Error during transcription: {e}")
        return

    # Step 3: Translate Text
    print("\nStep 3: Translating text...")
    try:
        translated_text = translate_text(original_text, target_language)
        print(f"Translated Text: {translated_text}")
    except Exception as e:
        print(f"Error during translation: {e}")
        return

    # Step 4: Text-to-Speech Conversion
    print("\nStep 4: Converting translated text to speech...")
    try:
        output_audio = "translated_audio.mp3"
        text_to_speech(translated_text, target_language, output_audio)
        print("Audio generation complete.")
    except Exception as e:
        print(f"Error during text-to-speech conversion: {e}")
        return

    # Step 5: Play the Audio
    print("\nStep 5: Playing translated audio...")
    display(Audio(output_audio, autoplay=True))

# Example usage
target_lang = "es"  # Set your target language code (e.g., 'es' for Spanish)
process_translation(target_lang)
```

```python
import gradio as gr
from gtts import gTTS
import os
import io
from pydub import AudioSegment
from googletrans import Translator
import speech_recognition as sr

# Initialize Translator and Recognizer
translator = Translator()
recognizer = sr.Recognizer()

def process_audio(audio_file, emotion, language):
    """
    Processes audio input: transcribes, translates, and generates audio output.
    """
    try:
        # Transcribe audio to text
        with sr.AudioFile(audio_file) as source:
            audio_data = recognizer.record(source)
            original_text = recognizer.recognize_google(audio_data)

        # Apply emotion (for now, just add a prefix to the text)
        emotion_text = f"[{emotion}] {original_text}"

        # Translate text to the chosen language
        translated_text = translator.translate(emotion_text, dest=language).text

        # Convert the translated text to speech
        tts = gTTS(translated_text, lang=language)
        audio_output = "output_audio.mp3"
        tts.save(audio_output)

        return translated_text, audio_output
    except Exception as e:
        return f"Error: {e}", None
def process_text(input_text, emotion, language):
    try:
        print(f"Input Text: {input_text}")
        print(f"Emotion: {emotion}")
        print(f"Target Language: {language}")

        # Apply emotion (for now, just add a prefix to the text)
        emotion_text = f"[{emotion}] {input_text}"

        # Translate text to the chosen language
        translated_text = translator.translate(emotion_text, dest=language).text
        print(f"Translated Text: {translated_text}")

        # Convert the translated text to speech
        tts = gTTS(translated_text, lang=language)
        audio_output = "output_audio.mp3"
        tts.save(audio_output)

        return translated_text, audio_output
    except Exception as e:
        print(f"Error: {e}")
        return f"Error: {e}", None


def create_interface():
    """
    Creates the Gradio interface for the application.
    """
    with gr.Blocks() as interface:
        gr.Markdown("## AI Voice Synthesis and Translation")

        with gr.Tab("Text Input"):
            input_text = gr.Textbox(label="Enter Text")
            emotion = gr.Dropdown(["Happy", "Sad", "Angry", "Neutral"], label="Select Emotion")
            language = gr.Textbox(label="Target Language Code (e.g., 'en', 'es', 'fr')")
            text_submit = gr.Button("Submit")
            text_output = gr.Textbox(label="Translated Text")
            text_audio = gr.Audio(label="Generated Audio", type="filepath", interactive=False)

        with gr.Tab("Audio Input"):
            audio_input = gr.Audio(type="filepath", label="Upload Audio")
            emotion_audio = gr.Dropdown(["Happy", "Sad", "Angry", "Neutral"], label="Select Emotion")
            language_audio = gr.Textbox(label="Target Language Code (e.g., 'en', 'es', 'fr')")
            audio_submit = gr.Button("Submit")
            audio_output = gr.Textbox(label="Translated Text")
            audio_audio = gr.Audio(label="Generated Audio", type="filepath", interactive=False)

        # Define functionality for the Text Input tab
        text_submit.click(
            process_text,
            inputs=[input_text, emotion, language],
            outputs=[text_output, text_audio]
        )

        # Define functionality for the Audio Input tab
        audio_submit.click(
            process_audio,
            inputs=[audio_input, emotion_audio, language_audio],
            outputs=[audio_output, audio_audio]
        )

    return interface

# Run the interface
interface = create_interface()
interface.launch()
```
